# Öncelikle bu hücreyi çalıştırın

```python
#Öncelikle bu hücreyi çalıştırın
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
veri = pd.read_csv("iris_csv.csv")
```

# **Niceliksel Özelliklerin Ortalamasının,Standart Sapmasının,Minimum ve Maksimmum Değerlerinin Hesaplanması**

**Sepal Uzunluk**

```python
sepal_length_toplami = veri["sepallength"].sum()
sepal_length_ortalamasi = veri["sepallength"].mean()
sepal_length_medyani = veri["sepallength"].median()
sepal_length_min = veri["sepallength"].min()
sepal_length_max = veri["sepallength"].max()
print("Toplam Sepal Uzunluğu:",sepal_length_toplami, "\nOrtalama Sepal Uzunluğu:", sepal_length_ortalamasi, "\nSepal Uzunluğunun Medyanı :",sepal_length_medyani,"\nSepal Uzunluğunun Minimum Değeri :",sepal_length_min,"\nSepal Uzunluğunun Maksimum Değeri :",sepal_length_max)
```

**Sepal Genişlik**

```python
sepal_width_toplami = veri["sepalwidth"].sum()
sepal_width_ortalamasi = veri["sepalwidth"].mean()
sepal_width_medyani = veri["sepalwidth"].median()
sepal_width_min = veri["sepalwidth"].min()
sepal_width_max = veri["sepalwidth"].max()
print("Toplam Sepal Genişliği:",sepal_width_toplami, "\nOrtalama Sepal Genişliği:", sepal_width_ortalamasi, "\nSepal Genişliğinin Medyanı:",sepal_width_medyani,"\nSepal Genişliğinin Minimum Değeri :",sepal_width_min,"\nSepal Genişliğinin Maksimum Değeri:",sepal_width_max )
```

**Petal Uzunluk**

```python
petal_length_toplami = veri["petallength"].sum()
petal_length_ortalamasi = veri["petallength"].mean()
petal_length_medyani = veri["petallength"].median()
petal_length_min = veri["petallength"].min()
petal_length_max = veri["petallength"].max()
print("Toplam Petal Uzunluğu:",petal_length_toplami, "\nOrtalama Petal Uzunluğu:", petal_length_ortalamasi, "\nPetal Uzunluğunun Medyanı :",petal_length_medyani,"\nPetal Uzunluğunun Minimum Değeri :",petal_length_min,"\nPetal Uzunluğunun Maksimum Değeri :",petal_length_max)
```

**Petal Genişlik**

```python
petal_width_toplami = veri["petalwidth"].sum()
petal_width_ortalamasi = veri["petalwidth"].mean()
petal_width_medyani = veri["petalwidth"].median()
petal_width_min = veri["petalwidth"].min()
petal_width_max = veri["petalwidth"].max()
print("Toplam Petal Genişliği:",petal_width_toplami, "\nOrtalama Petal Genişliği:", petal_width_ortalamasi, "\nPetal Genişliğinin Medyanı:",petal_width_medyani,"\nPetal Genişliğinin Minimum Değeri :",petal_width_min,"\nPetal Genişliğinin Maksimum Değeri:",petal_width_max )
```

# **Kovaryans Ve Korelasyon Hesabı**

**Kovaryans Hesabı Ve Grafiksel Çizimi**

<center><img src="https://kredi-hesaplama.com/wp-content/uploads/2017/07/Kovaryans-ve-Korelasyon-Nedir.png", width='500px'></center>
<br>




```python
sepal_length = veri["sepallength"]
sepal_width = veri["sepalwidth"]
petal_length = veri["petallength"]
petal_width = veri["petalwidth"]

iris_cov = np.array([sepal_length,sepal_width,petal_length,petal_width])

covMatrix = np.cov(iris_cov,bias=True)        #Matrisi çizdirdim.
print(covMatrix)
sn.heatmap(covMatrix, annot=True, fmt='g')
plt.show()
```

**Korelasyon Hesabı Ve Grafiksel Çizimi**
<center><img src="https://www.statisticshowto.com/wp-content/uploads/2012/10/pearson.gif", width='450px'></center>

```python
sepal_length = veri["sepallength"]
sepal_width = veri["sepalwidth"]
petal_length = veri["petallength"]
petal_width = veri["petalwidth"]

iris_corrcoef = np.array([sepal_length,sepal_width,petal_length,petal_width])

iris_corrcoef_result = np.corrcoef(iris_corrcoef)
print(iris_corrcoef_result)

sn.heatmap(iris_corrcoef_result, annot=True, fmt='g')
plt.show()
```

# **Histogram Çizimi**

```python
fig = plt.figure(figsize=(10,5))

plt.subplot(221)
plt.hist(sepal_length, bins = 8,density=True,histtype='bar')
plt.xlabel('Değer',fontsize=10);plt.ylabel('Frekans',fontsize=10)
plt.title('Sepal Uzunluk')

plt.subplot(222)
plt.hist(sepal_width, bins = 8,density=True,histtype='bar')
plt.xlabel('Değer',fontsize=10);plt.ylabel('Frekans',fontsize=10)
plt.title('Sepal Genişlik')

plt.subplot(223)
plt.hist(petal_length, bins = 8,density=True,histtype='bar')
plt.xlabel('Değer',fontsize=10);plt.ylabel('Frekans',fontsize=10)
plt.title('Petal Uzunluk')

plt.subplot(224)
plt.hist(petal_width, bins = 8,density=True,histtype='bar')
plt.xlabel('Değer',fontsize=10);plt.ylabel('Frekans',fontsize=10)
plt.title('Petal Genişlik')



fig.tight_layout(rect=(0,0,2,2))
plt.show()
```
